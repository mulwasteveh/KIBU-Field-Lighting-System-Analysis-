# Analysing latitude trend based on the kibabii field lighting system

```python
data=pd.read_csv("D:\ML datasets\GIS Data Collection.csv")

df = pd.DataFrame(data)
data.head(5)
```

```python
X = list(df.iloc[:, 1])
Y = list(df.iloc[:, 2])
# Plot the data using plot() method. That is the line graph.
plt.plot(Y, X)
plt.title("Bar graph showing different lighting posts by latitude at which they lie")
plt.xlabel("Latitudes")
plt.ylabel("Lighting posts around kibabii university field")
plt.rcParams["figure.figsize"] = (20,20)
plt.grid(True)
  
# Show the plot
plt.show()
```

# Analysing longitudes trend based on the kibabii field lighting system

```python
X = list(df.iloc[:, 1])
Y = list(df.iloc[:, 3])

# Plot the data using bar() method
plt.plot(Y, X)
plt.title("Bar graph showing different lighting posts by longitude at which they lie")
plt.xlabel("Longitudes")
plt.ylabel("Lighting posts around kibabii university field")
plt.rcParams["figure.figsize"] = (20,10)
plt.grid(True)
  
# Show the plot
plt.show()
```
